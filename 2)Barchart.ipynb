```python
# a problem:- Over lapping the names on x axes
# simple bar chart
children=[10,20,30,50,25]
color=['red red red red ',' blue blue blue blue ',' orange orange orange  ', 'green green green ','white white white']
plt.bar(color,children,color='#0BAAF4')

```

```python
# Solution:-xtick()
children=[10,20,30,50,25]
color=['red red red red ',' blue blue blue ','orrange orrange Orrangw  ', 'green green green  ',' white white white']
plt.bar(color,children,color='#0BAAF4')

plt.xticks(rotation='vertical')

```

```python
# Stacked Bar chart
#Problem:- Overlap in Staked bar Chart
plt.bar(batsman['batsman'],batsman['2017'])
plt.bar(batsman['batsman'],batsman['2016'])
plt.bar(batsman['batsman'],batsman['2015'])
```

```python
# Stacked Bar chart
#Problem:- Overlap in Stacked bar Chart
plt.bar(batsman['batsman'], batsman['2017'])
plt.bar(batsman['batsman'], batsman['2016'], bottom=batsman['2017'])
plt.bar(batsman['batsman'], batsman['2015'], bottom=(batsman['2016'] + batsman['2017']))
```
