```python
# Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Colore Scatterp plots

```python
iris=pd.read_csv(r'iris.csv')
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-versicolor':0,'Iris-virginica':1,'Iris-setosa':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter') # c= is a hidden parameter color # all points red
plt.xlabel('Sepal Length')                                                               # cmap is used to set the theme
plt.ylabel('Petal Length')
```

```python
#colbar /alpha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.3) # c= is a hidden parameter # all points red
plt.xlabel('Sepal Length')                                                               # cmap is used to set the theme
plt.ylabel('Petal Length')
plt.colorbar()
```

# Plot size

```python
plt.figure(figsize=(12,5))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.3) # c= is a hidden parameter # all points red
plt.xlabel('Sepal Length')                                                               # cmap is used to set the theme
plt.ylabel('Petal Length')
plt.colorbar()
```

# Annotations 

```python
batters=pd.read_csv(r'batter.csv')
```

```python
batters.head(5)
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.figure(figsize=(18,10))  # set figure size first

plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs']) # s for size

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
```

```python
# text() / fontdict{}
x=[1,2,3,4]
y=[5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Pointer:1')
plt.text(2,6,'Pointer:2')
plt.text(3,7,'Pointer:3',fontdict={'size':22,'color':'Brown'})
plt.text(4,8,'Pointer:4')
```

# Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))  # set figure size first

plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs']) # s for size

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
plt.axhline(130,c='red')
plt.axvline(30,c='green')

```

```python
# SubPlots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6),sharex=True) #,sharex=True It helps to off the not show the coordinates

ax[0].scatter(batters['avg'], batters['strike_rate'],c='red')
ax[0].set_title("Avg vs Strike Rate")
ax[0].set_ylabel('strike_rate')

ax[1].scatter(batters['avg'], batters['runs'],c='#F4780B')
ax[1].set_title("Avg vs Runs")
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
plt.show()
```

```python
ax[0]
```

```python
# SubPlots 4 graphs
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

ax[0,0].scatter(batters['avg'], batters['strike_rate'],c='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```
