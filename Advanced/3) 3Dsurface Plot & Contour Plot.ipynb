```python
# Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 3d Surface plot 

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x,y=np.meshgrid(x,y)
```

```python
x.shape
```

```python
z=x**2 + y**2
z.shape
```

```python
fig= plt.figure(figsize=(9,6))

ax=plt.subplot(projection='3d')
main=ax.plot_surface(x,y,z, cmap='viridis')
fig.colorbar(main)
```

```python
z=np.sin(x) +np.cos(y)
fig= plt.figure(figsize=(9,6))

ax=plt.subplot(projection='3d')
main=ax.plot_surface(x,y,z, cmap='viridis')
fig.colorbar(main)
```

# Contour Plot 

```python
# It is used to show the 3d graph in 2d
```

```python
fig= plt.figure(figsize=(9,6))

ax=plt.subplot()

main=ax.contour(x,y,z, cmap='viridis')
fig.colorbar(main)
```

```python
fig= plt.figure(figsize=(9,6))

ax=plt.subplot()

main=ax.contourf(x,y,z, cmap='viridis')
fig.colorbar(main)
```

```python
z=np.sin(x) +np.cos(y)
fig= plt.figure(figsize=(9,6))

ax=plt.subplot()
main=ax.contourf(x,y,z, cmap='viridis')
fig.colorbar(main)
```
