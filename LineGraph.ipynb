```python
# Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
# from a pandas dataframe
batsman=pd.read_csv(r'sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
#colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE')

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# legend -> locationplt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE',#linestyle='dotted'
        linestyle='dashdot')

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
#Line width
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3.5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE',linestyle='dashdot',linewidth=1)

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# Marker/ markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3.5,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE',linestyle='dashdot',linewidth=1,marker='+',markersize=25)

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# legend -> Location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3.5,marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE',linestyle='dashdot',linewidth=1,marker='+',markersize=25,label='Rohit')

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend() ## For label's
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3.5,marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF00AE',linestyle='dashdot',linewidth=1,marker='+',markersize=25,label='Rohit')

plt.title("Rohit Sharma V/s Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend() ## For label's
plt.legend(loc='upper right')
# best
# upper right
# upper left
# lower right
# lower left
# right
# center left
# center right
# lower center
# upper center
# center
```
