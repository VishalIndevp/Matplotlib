```python
# Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
x=np.linspace(-10,10,50)
# x
y=10*10 +3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# pit. scatter on pandas data
df=pd.read_csv(r'batter.csv')
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='#00D9FF',marker='*')
plt.title("Avg and SR Analysis of top 50 batsman")
plt.xlabel("Average")
plt.ylabel("Strike_rate")
```
